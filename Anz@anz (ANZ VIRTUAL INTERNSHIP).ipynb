```python
## Find mode of card_present_flag column to fill misssing value
data1['card_present_flag'].mode()
```

```python
result = data1[data1['txn_description'] == 'PAY/SALARY'].groupby('account')['amount'].sum()
result.head()
```

```python
## Graph of salary distribution
result.hist(bins=20)
plt.xlabel('Salary Distribution')
plt.show()
```

```python
acc_grp = data1.groupby('account')
##quantity_ordered = product_group.sum()['Quantity Ordered']

accounts = [product for product, data1 in acc_grp]
plt.bar(accounts, result)
plt.xticks(rotation = 'vertical')
plt.show()
```
